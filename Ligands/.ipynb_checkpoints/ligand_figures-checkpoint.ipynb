```python
from rdkit import Chem
from rdkit.Chem import AllChem
from collections import defaultdict
from rdkit.Chem import rdFMCS
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole

#needed for show_mols
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import SVG
import cairosvg
import math
import os

from rdkit import rdBase
```

```python
#inspired by https://github.com/rdkit/UGM_2020/blob/master/Notebooks/Landrum_WhatsNew.ipynb
def show_mols(mols, mols_per_row = 5, size=200, min_font_size=12, legends=[], file_name=''):
    if legends and len(legends) < len(mols):
        print('legends is too short')
        return None

    mols_per_row = min(len(mols), mols_per_row)  
    rows = math.ceil(len(mols)/mols_per_row)
    d2d = rdMolDraw2D.MolDraw2DSVG(mols_per_row*size,rows*size,size,size)
    
    d2d.drawOptions().minFontSize = min_font_size
    d2d.drawOptions().SetFontSize = min_font_size
    d2d.drawOptions().maxFontSize = -1
    
    

    
#    d2d.setScale

    if legends:
        d2d.DrawMolecules(mols, legends=legends)
        #d2d.drawOptions().legendFontSize = 100000
        d2d.drawOptions().lengedFraction = 0.9
    else:
        d2d.DrawMolecules(mols)
    d2d.FinishDrawing()

    if file_name:
        with open('d2d.svg', 'w') as f:
            f.write(d2d.GetDrawingText())
        if 'pdf' in file_name:
            cairosvg.svg2pdf(url='d2d.svg', write_to=file_name)
        else:
            cairosvg.svg2png(url='d2d.svg', write_to=file_name, dpi=1200, scale=2)
        #os.remove('d2d.svg')

    return d2d, SVG(d2d.GetDrawingText())
```

```python
smis = []
numbers = []
with open("mup1_ligands.txt", 'r') as fi:
    for line in fi.readlines():
        split = line.split()
        smis.append(split[0])
        numbers.append(split[2])
        
names = ["SBT-S", "SBT-R", "IBT", "PT", "IPT", "ET", "MT", "PE9", "HE2", "HE4", "OC9", "ODI", "PRZ", "IPZ"]

mols = [Chem.MolFromSmiles(smi) for smi in smis]

legends = []
for i, name in enumerate(names):
    legends.append(f"{numbers[i]}: {name}")
    
x, _ = show_mols(mols, mols_per_row=7, size=250, min_font_size=30, legends=legends, file_name='mup1_mols_png.png')

```
